```python
Savis_Books.loc[Savis_Books['Series_ID']==4]
```

```python
#.loc[[]==] creates filter for data
```

```python
Savis_Books.loc[(Savis_Books['Series_ID']==4)&(Savis_Books['Release_Date']>=2009)]
```

```python
#.loc[([]==)&([]>=)] creates multiple filters
```

```python
Blue_Bloods_Series=Savis_Books.loc[Savis_Books['Series_ID']==4]
```

```python
#Names and creates new set of data from the filters
```

```python
Blue_Bloods_Series
```

```python
Blue_Bloods_Series.sort_values(by=['Release_Date'],axis=0)
```

```python
#.sort_values(by=[],axis=0) puts data in order you choose
```

```python
Blue_Bloods_Series.to_csv('Blue Bloods Series.csv',index=False)
```

```python
#.to_csv('',index=False) turns your table into a real file
#index=False keeps the random weird numbers away
```
